```python
import matplotlib.pyplot as plt

import numpy as np

import os
import sys

sys.path.append('..')
import data
import utils

%matplotlib inline
```

```python
params = utils.yaml_to_dict(os.path.join('..','config.yml'))
params['data_dir'] = os.path.join('..',params['data_dir'])
params['model_dir'] = os.path.join('..',params['model_dir'])
params['batch_size'] = 50
```

```python
print('Data in train')
train_generator = data.make_datagenerator(params, mode='training')
```

```python
data_in_generator = next(train_generator)

images = data_in_generator[0]
labels = data_in_generator[1]
```

```python
id_label_map = utils.load_id_label_map(params)
    
fig, axis = plt.subplots(15,3, figsize=(20,30))

for i in range(15):
    k = np.random.randint(0,50)
    
    label = id_label_map[labels[k]]
    axis[i][0].text(0.5, 0.5, label, size=25, ha="center", va="center",
            bbox=dict(
                boxstyle="round",
                ec=(1., 0.5, 0.5),
                fc=(1., 0.8, 0.8),
            )
           )
    axis[i][0].axis('off')
    
    axis[i][1].imshow((images[k,:,0:80,:]*255).astype(int))
    axis[i][1].axis('off')
    
    axis[i][2].imshow((images[k,:,0:80,:]*255).astype(int))
    axis[i][2].axis('off')

plt.show()
```

```python
id_label_map = utils.load_id_label_map(params)
fig, axis = plt.subplots(5,5, figsize=(20,30))

for i in range(5):
    for j in range(5):
        
        k = np.random.randint(0,50)
        
        axis[i][j].imshow((images[k,:,0:80,:]*255).astype(int))
        axis[i][0].axis('off')
        
        axis[i][j].set_title(id_label_map[labels[k]])
        axis[i][j].axis('off')
        
plt.show()
```
